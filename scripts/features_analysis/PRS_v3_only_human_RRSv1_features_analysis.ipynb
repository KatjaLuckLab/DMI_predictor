```python
elm_class_df= elm_df[['ELM name']].copy()
elm_class_df.drop_duplicates(inplace= True)
elm_class_df['Class']= elm_class_df['ELM name'].str.slice(stop= 3)
DMI_count_df= pd.DataFrame(data= {'Class': list(elm_class_df.groupby('Class').count().index), 'ElmDB': [v[0] for v in elm_class_df.groupby('Class').count().values]})
DMI_count_df
```

```python
PRSv3_DMI_count_df= pd.DataFrame(PRSv3.Elm.unique())
PRSv3_DMI_count_df= PRSv3_DMI_count_df.rename(columns= {0: 'Elm'})
PRSv3_DMI_count_df['Class']= PRSv3_DMI_count_df.Elm.str.slice(stop=3)
PRSv3_DMI_vc_df= pd.DataFrame(data= {'Class': PRSv3_DMI_count_df.Class.value_counts().index, 'PRSv3': PRSv3_DMI_count_df.Class.value_counts().values})
PRSv3_DMI_vc_df
```

```python
DMI_count_df= DMI_count_df.merge(PRSv3_DMI_vc_df, how= 'inner')
```

```python
DMI_count_df
```

```python
RRSv1_DMI_count_df= pd.DataFrame(RRSv1.Elm.unique())
RRSv1_DMI_count_df= RRSv1_DMI_count_df.rename(columns= {0: 'Elm'})
RRSv1_DMI_count_df['Class']= RRSv1_DMI_count_df.Elm.str.slice(stop=3)
RRSv1_DMI_vc_df= pd.DataFrame(data= {'Class': RRSv1_DMI_count_df.Class.value_counts().index, 'RRSv1': RRSv1_DMI_count_df.Class.value_counts().values})
RRSv1_DMI_vc_df
```

```python
DMI_count_df= DMI_count_df.merge(RRSv1_DMI_vc_df)
```

```python
DMI_count_df
```

```python
DMI_count_df['PRSv3_fraction']= DMI_count_df['PRSv3']/DMI_count_df['ElmDB']
DMI_count_df['RRSv1_fraction']= DMI_count_df['RRSv1']/DMI_count_df['ElmDB']
DMI_count_df
```

```python
N= 6
ind= np.arange(N)
width= 0.3

plt.figure(figsize= (8,6))
plt.bar(ind, DMI_count_df.PRSv3_fraction, width, color= 'c', label= 'PRS')
plt.bar(ind + width, DMI_count_df.RRSv1_fraction, width, color= 'b', label= 'RRSv1')

plt.xticks(ind + width / 2, DMI_count_df.Class)
plt.title('Fraction of DMI represented in the PRSv3 and RRSv1 by class')
plt.ylabel('Fraction of DMI types with DMI instance')
plt.legend(loc= 'best')

plt.tight_layout()
plt.savefig('/Users/chopyanlee/Coding/Python/DMI/PRSv3_RRSv1_analysis_plots/DMI_fraction_PRSv3_RRSv1.pdf')

plt.show()
```

```python
fig, (ax1, ax2)= plt.subplots(1, 2, figsize= (12, 6), sharey= True)

ax1.bar(DMI_count_df.Class, DMI_count_df.PRSv3_fraction)
ax1.set_title('Fraction of DMI represented in the PRSv3 by class')

ax2.bar(DMI_count_df.Class, DMI_count_df.RRSv1_fraction)
ax2.set_title('Fraction of DMI represented in the RRSv1 by class')

ax1.set_ylabel('Fraction of DMI types with DMI instance')

plt.show()
```

While the representation of DMI types by class in PRSv3 is low and uneven in general, that of RRSv1 is much higher in comparison.

```python
PRSv3_evalues= []
for i, r in PRSv3.iterrows():
    ev= ''
    ev2= ''
    if type(r['DomainMatchEvalue1']) == str:
        ev= r['DomainMatchEvalue1'].split('|')
        for e in ev:
            PRSv3_evalues.append(float(e))
    else:
        PRSv3_evalues.append(float(r['DomainMatchEvalue1']))
    if pd.isna(r['DomainMatchEvalue2']) == False:
        if type(r['DomainMatchEvalue2']) == str:
            ev2= r['DomainMatchEvalue2'].split('|')
            for e2 in ev2:
                PRSv3_evalues.append(float(e2))
        else:
            PRSv3_evalues.append(float(r['DomainMatchEvalue2']))
len(PRSv3_evalues)
```

```python
RRSv1_evalues= []
for i, r in RRSv1.iterrows():
    ev= ''
    ev2= ''
    if type(r['DomainMatchEvalue1']) == str:
        ev= r['DomainMatchEvalue1'].split('|')
        for e in ev:
            RRSv1_evalues.append(float(e))
    else:
        RRSv1_evalues.append(float(r['DomainMatchEvalue1']))
    if pd.isna(r['DomainMatchEvalue2']) == False:
        if type(r['DomainMatchEvalue2']) == str:
            ev2= r['DomainMatchEvalue2'].split('|')
            for e2 in ev2:
                RRSv1_evalues.append(float(e2))
        else:
            RRSv1_evalues.append(float(r['DomainMatchEvalue2']))
len(RRSv1_evalues)
```

```python
b= np.logspace(np.log10(min([ev for ev in PRSv3_evalues + RRSv1_evalues if ev != 0])), np.log10(max(PRSv3_evalues + RRSv1_evalues)), 100)

fig, (ax1, ax2)= plt.subplots(2, 1, figsize= (8,9), sharex= True)

ax1.hist(PRSv3_evalues, bins= b)
ax1.set_title('Distribution of domain match e-values in PRSv3')
ax1.set_xscale('log')

ax2.hist(RRSv1_evalues, bins= b)
ax2.set_title('Distribution of domain match e-values in RRSv1')
ax2.set_xscale('log')

fig.text(0.06, 0.5, 'Number of domain match', ha='center', va='center', rotation='vertical')

plt.savefig('/Users/chopyanlee/Coding/Python/DMI/PRSv3_RRSv1_analysis_plots/evalues_PRSv3_RRSv1.pdf')

plt.show()
```

The distribution of evalues of domain matches in PRS and RRSv1 are both skewed to the right, but that of PRS has more data points scattered to the left.
